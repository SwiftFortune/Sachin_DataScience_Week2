### Task 2: Working with Lists and Arrays 

##### 1. Create a list containing at least 10 numbers. 

```python
# Create a list of containing at least 10 numbers
numbers = [12, 45, 23, 67, 34, 89, 90, 11, 5, 78]
```

##### 2. Perform the following: 

```python
# Print the length of the list
print(f"Length of the list: {len(numbers)}")
# find the maximun and mininum number in the list
print(f"Maximun number: {max(numbers)}")
print(f"Mininum number: {min(numbers)}")
# Add a new element and remove an element from the list
numbers.append(100)
print(f"List after adding 100: {numbers}")
numbers.remove(23)
print(f"List after removing 23: {numbers}")
# Sort the list in ascending and descending order
numbers.sort()
print(f"List in ascending order: {numbers}")
numbers.sort(reverse=True)
print(f"List in descending order: {numbers}")
```

### 3. Convert the list into a NumPy array and calculate: 

```python
# Convert the list into a Numpy array 
import numpy as np
np_array = np.array(numbers)
print(f"Numpy array: {np_array}")

index_min = np.argmin(np_array)
print(f"Index of minimun element: {index_min}")
```

```python
# Calculate the following statistics from the list:
# Mean
mean = np.mean(np_array)
print(f"Mean: {mean}")
# Median
median = np.median(np_array)
print(f"Median: {median}")
# Standard Deviation
std_dev = np.std(np_array)
print(f"Standard Deviation: {std_dev}")
# Variance 
variance = np.var(np_array)
print(f"Variancd: {variance}")
# Sum of all elements 
sum_elements = np.sum(np_array)
print(f"Sum of all elements: {sum_elements}")
# Count of elements greater than a specified value 
count_greater_than_50 = np.sum(np_array > 50)
print(f"Count of elements greater than 50: {count_greater_than_50}")
# Find the index of the maximun and minimun elements
index_max = np.argmax(np_array)
print(f"Index of maximun element: {index_max}")
```

##### 5. Perform set operations — union, intersection, and difference — between two sets of sample data.

```python
# Perform set operations — union, intersection, and difference — between two sets of sample data.
set1 = {"Python", "Data Science", "AI"}
set2 = {"AI", "Machine Learning", "Deep Learning"}
# Union 
union_set = set1.union(set2)
print(f"Union: {union_set}")
# Intersection
intersection_set = set1.intersection(set2)
print(f"Intersection: {intersection_set}")
# Difference
difference_set = set1.difference(set2)
print(f"Difference: {difference_set}")
```

##### 2. Write the following information into the file (name, course, and marks of at least 5 students). 
##### 3. Read the file and display only those students whose marks are above 75.

```python
# Step 1 & 2: Write student information into a file
with open("student_data.txt", "w") as file:
    students = [
        {"name": "Alice", "course": "Math", "marks": 85},
        {"name": "Bob", "course": "Physics", "marks": 90},
        {"name": "Charlie", "course": "Chemistry", "marks": 78},
        {"name": "David", "course": "Biology", "marks": 92},
        {"name": "Eve", "course": "Computer Science", "marks": 88}
    ]
    
    for stu in students:
        file.write(f"Name: {stu['name']}, Course: {stu['course']}, Marks: {stu['marks']}\n")


# Step 3: Read file and display only students whose marks are above 75
print("Students with marks above 75:")
with open("student_data.txt", "r") as file:
    for line in file:
        line = line.strip()

        # Skip empty or malformed lines
        if not line or "Marks:" not in line:
            continue
        
        parts = line.split(", ")

        # Extract marks
        marks = int(parts[2].split(": ")[1])

        if marks > 75:
            print(line)
```
